```python
# 2017 open close (  )    

# (/()) -> ROS  
## () = (+)) 
###    ()
###      
###      .


import csv
import matplotlib.pyplot as plt

f = open('train.csv')
data = csv.reader(f)
next(data)
open_ROS = []
close_ROS = []


for row in data:
    try:
        aocnf = int(row[7])  #
        extra = int(row[11])  #
        tnsdldlr = int(row[15])  #
        ROS = (tnsdldlr)/(aocnf+extra) 

        if row[1] == 'open' :
            open_ROS.append(ROS)

        else:
            close_ROS.append(ROS)

    except:
        pass
plt.title('2017   ')
plt.plot(open_ROS,'b',label = 'open')
plt.plot(close_ROS,'r', label = 'close')
plt.ylim([-1.5,1.5])
plt.legend()
plt.show()
```

```python
#2017    (  ,    )
# = [7]revenue - [8]salescost - [9]sga
##     


import csv
import matplotlib.pyplot as plt

f = open('train.csv')
data = csv.reader(f)
next(data)
open_ROS = []
close_ROS = []


for row in data:
    try:
        aocnf = int(row[7])
        aocnfdnjsrk = int(row[8])
        vksaorhksflql = int(row[9])
        duddjqdldlr = aocnf-aocnfdnjsrk-vksaorhksflql
        ratio = duddjqdldlr/aocnf

        if row[1] == 'open' :
            open_ROS.append(ratio)

        else:
            close_ROS.append(ratio)
    except:
        pass
plt.title('2017   ')
plt.plot(open_ROS,'b',label = 'open')
plt.plot(close_ROS,'r', label = 'close')
plt.ylim([-1.5,1.5])
plt.legend()
plt.show()
```

```python
#
# = 7revenue - 8salescost - 9sga
#2017   
#      

import csv
import matplotlib.pyplot as plt

f = open('train.csv')
data = csv.reader(f)
next(data)
open_ratio = []
close_ratio = []
label = ['','']
for row in data:
    if row[1] == 'open' :
        try:
            aocnf = int(row[7])
            aocnfdnjsrk = int(row[8])
            vksaorhksflql = int(row[9])
            duddjqdldlr = aocnf-aocnfdnjsrk-vksaorhksflql
#            print(aocnf, aocnfdnjsrk, vksaorhksflql, duddjqdldlr)
            open_ratio.append(aocnf)
            open_ratio.append(duddjqdldlr)
            break
   
        except:
            pass

plt.title('2017 open sample1   ')        
plt.rc('font',family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.axis('equal')
plt.pie(open_ratio, labels=label, autopct = '%.1f%%')
plt.legend()
plt.show()       

```
