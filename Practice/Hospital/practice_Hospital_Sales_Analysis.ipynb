```python
#2017매출액과 순이익 한 그래프에 나타내기
import csv
import matplotlib.pyplot as plt

f = open('train.csv')
data = csv.reader(f)
next(data)
open_revenue1 = []
open_profit1 = []
close_revenue1 = []
close_profit1 = []

for row in data:
    try:
        int(row[7])  #매출액 숫자로
        int(row[15])  #순이익 숫자로
        if row[1] == 'open':
            open_revenue1.append(row[7])  #오픈매출액추가
            open_profit1.append(row[15])  #오픈순이익추가

        else:
            close_revenue1.append(row[7])  #close매출액추가
            close_profit1.append(row[15])  #close순이익추가
    except:
        pass
    
#open병원 매출액과 순이익 한 그래프에 나타내기
plt.plot(open_revenue1,'b', label = '매출액')  
plt.plot(open_profit1,'r', label = '순이익')
plt.title('Open병원의 2017 매출액과 순이익')
plt.legend() 
plt.show()
#close병원 매출액과 순이익 한 그래프에 나타내기
plt.plot(close_revenue1,'b', label = '매출액')
plt.plot(close_profit1,'r', label = '순이익')
plt.title('Close병원의 2017 매출액과 순이익')
plt.legend()
plt.show()

#open병원은 매출액의 산포가 커서 축 눈금이 안보임...
```

```python
# 2017년도 open과 close의 (총수익과 순이익의 비율)을 한 그래프에 나타내고 비교

# (순이익/(총수익)) -> ROS로 표현했다 
## (총수익) = (매출+영업외수익)) 
### 매출은 영업으로만 벌어들은 수익(영업외수익비포함)이고
### 순이익은 영업외 순수익까지 포함한 항목이므로 
### 매출과 영업외수익을 더해 총 수익을 구했다.


import csv
import matplotlib.pyplot as plt

f = open('train.csv')
data = csv.reader(f)
next(data)
open_ROS = []
close_ROS = []


for row in data:
    try:
        aocnf = int(row[7])  #매출
        extra = int(row[11])  #영업외수익
        tnsdldlr = int(row[15])  #순이익
        ROS = (tnsdldlr)/(aocnf+extra) 

        if row[1] == 'open' :
            open_ROS.append(ROS)

        else:
            close_ROS.append(ROS)

    except:
        pass
plt.title('2017년도 총수익과 순이익의 비율')
plt.plot(open_ROS,'b',label = 'open')
plt.plot(close_ROS,'r', label = 'close')
plt.ylim([-1.5,1.5])
plt.legend()
plt.show()
```

```python
#2016 open과 close의 (총수익과 순이익의 비율)을 한 그래프에 나타내고 비교
#년도만 다른 위와 같은 그래프.
import csv
import matplotlib.pyplot as plt

f = open('train.csv')
data = csv.reader(f)
next(data)
open_ROS = []
close_ROS = []


for row in data:
    try:
        aocnf = int(row[31])
        extra = int(row[35])
        tnsdldlr = int(row[39])
        ROS = (tnsdldlr)/(aocnf+extra)

        if row[1] == 'open' :
            open_ROS.append(ROS)

        else:
            close_ROS.append(ROS)

    except:
        pass
plt.title('2016년도 총수익과 순이익의 비율')
plt.plot(open_ROS,'b', label = 'open')
plt.plot(close_ROS,'r',label = 'close')
plt.ylim([-1.5,1])
plt.legend()
plt.show()
```

```python
#2017 매출 대비 영업이익 (매출은 영업이익과 매출원가, 판매비와 관리비를 합친 값)
#영업이익 = [7]매출revenue - [8]매출원가salescost - [9]판매비와관리비sga
##영업이익값이 표에 없으므로 위 계산식으로 구했다


import csv
import matplotlib.pyplot as plt

f = open('train.csv')
data = csv.reader(f)
next(data)
open_ROS = []
close_ROS = []


for row in data:
    try:
        aocnf = int(row[7])
        aocnfdnjsrk = int(row[8])
        vksaorhksflql = int(row[9])
        duddjqdldlr = aocnf-aocnfdnjsrk-vksaorhksflql
        ratio = duddjqdldlr/aocnf

        if row[1] == 'open' :
            open_ROS.append(ratio)

        else:
            close_ROS.append(ratio)
    except:
        pass
plt.title('2017 매출 대비 영업이익')
plt.plot(open_ROS,'b',label = 'open')
plt.plot(close_ROS,'r', label = 'close')
plt.ylim([-1.5,1.5])
plt.legend()
plt.show()
```

```python
#2016 매출 대비 영업이익
#영업이익 = [31]매출revenue - [32]매출원가salescost - [33]판매비와관리비sga
#년도만 다른 위와 같은 그래프

import csv
import matplotlib.pyplot as plt

f = open('train.csv')
data = csv.reader(f)
next(data)
open_ROS = []
close_ROS = []


for row in data:
    try:
        aocnf = int(row[31])
        aocnfdnjsrk = int(row[32])
        vksaorhksflql = int(row[33])
        duddjqdldlr = aocnf-aocnfdnjsrk-vksaorhksflql
        ratio = duddjqdldlr/aocnf

        if row[1] == 'open' :
            open_ROS.append(ratio)

        else:
            close_ROS.append(ratio)

    except:
        pass
plt.title('2016 매출 대비 영업이익')
plt.plot(open_ROS,'b', label = 'open')
plt.plot(close_ROS,'r', label = 'close')
plt.ylim([-1.5,1.5])
plt.legend()
plt.show()
```

```python
#파이차트
#영업이익 = 7매출revenue - 8매출원가salescost - 9판매비와관리비sga
#2017 매출 대비 영업이익
#한글 출력 및 음수 출력시 문제 해결

import csv
import matplotlib.pyplot as plt

f = open('train.csv')
data = csv.reader(f)
next(data)
open_ratio = []
close_ratio = []
label = ['매출','영업이익']
for row in data:
    if row[1] == 'open' :
        try:
            aocnf = int(row[7])
            aocnfdnjsrk = int(row[8])
            vksaorhksflql = int(row[9])
            duddjqdldlr = aocnf-aocnfdnjsrk-vksaorhksflql
#            print(aocnf, aocnfdnjsrk, vksaorhksflql, duddjqdldlr)
            open_ratio.append(aocnf)
            open_ratio.append(duddjqdldlr)
            break
   
        except:
            pass

plt.title('2017 open sample1 매출 대비 영업이익')        
plt.rc('font',family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.axis('equal')
plt.pie(open_ratio, labels=label, autopct = '%.1f%%')
plt.legend()
plt.show()       

```
