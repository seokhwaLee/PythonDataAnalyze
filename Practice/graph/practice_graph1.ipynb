```python
#8월의 최고기온 데이터 시각화
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '08' :
            result.append(float(row[-1]))
plt.plot(result, 'r')
plt.show()
```

```python
#내 생일(1997.05.21)의 태어난 이후 매년 최고기온 시각화
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[-1] != '' :
        if 1997 <= int(row[0].split('-')[0]) :
            if row[0].split('-')[1] == '05' and row[0].split('-')[2] == '21':
                result.append(float(row[-1]))
plt.plot(result, 'hotpink')
plt.show()
```

```python
#내 생일(1997.05.21)의 태어난 이후 매년 최고기온 시각화
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
hot = []
cold = []

for row in data :
    if row[-1] != '' :
        if 1997 <= int(row[0].split('-')[0]) :
            if row[0].split('-')[1] == '05' and row[0].split('-')[2] == '21':
                hot.append(float(row[-1]))
                cold.append(float(row[-2]))
                                  
plt.plot(hot, 'hotpink', label='hot')
plt.plot(cold, 'blue', label = 'cold')
plt.legend()
plt.show()
```
