```python
import csv
f=open('age1.csv')
data = csv.reader(f)
result = []
for row in data:
    if '신도림' in row[0]:
        for i in row[1]:
            result.append(int(i))
import matplotlib.pyplot as plt
plt.bar(range(101),result)
plt.show()

##그래프 그리려고 했는데 숫자에 ','가 있어서 오류뜸......
```

```python
# 스플릿을 사용해 숫자안 ','제거...
import csv
f=open('age1.csv')
data = csv.reader(f)
result = []

for row in data:
    if '구로' in row[0]:
        row1 = row[1].split(',')
        row1=''.join(row1)
        result.append(int(row1))

import matplotlib.pyplot as plt
plt.title('2019년 2월 구로구 총인구수,동별인구수')
plt.bar(range(16),result)
plt.show()
```
